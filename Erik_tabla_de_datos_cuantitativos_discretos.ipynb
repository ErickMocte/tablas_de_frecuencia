```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de Seaborn
sns.set(style="whitegrid")

# Generar 500 datos aleatorios enteros entre 0 y 100
data = np.random.randint(0, 101, 500)

# Crear un DataFrame para facilitar el manejo de datos
df = pd.DataFrame(data, columns=['Valores'])

# Mostrar la tabla de datos generados
tabla_resumen = df.describe().round(2)
print("Tabla de datos generados:")
print(tabla_resumen)

# Gráfico de frecuencia absoluta (histograma)
plt.figure(figsize=(10, 6))
plt.hist(df['Valores'], bins=10, color='slategray', alpha=0.8)  # Color más profesional
plt.title('Histograma de Conteo de Ocurrencias', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Conteo de Ocurrencias', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(bottom=0)
plt.show()

# Gráfico de barras horizontales (simplificado)
frecuencia = df['Valores'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
frecuencia[:10].plot(kind='barh', color='darkorange', alpha=0.8)  # Color más profesional
plt.title('Diagrama de Barras Horizontales (Top 10 Valores)', fontsize=16)
plt.xlabel('Número de Ocurrencias', fontsize=14)
plt.ylabel('Rangos de Valores', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Gráfico de frecuencia relativa (polígono de frecuencia)
frecuencia_relativa = frecuencia / len(df)
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_relativa.index, frecuencia_relativa.values, marker='o', linestyle='-', color='teal', markersize=8)  # Color más profesional
plt.title('Polígono de Proporción de Datos', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Proporción de Datos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.show()

# Diagrama de pastel
plt.figure(figsize=(10, 6))
colores = ['lightgrey'] * len(frecuencia[:10])
max_index = frecuencia[:10].idxmax()
colores[max_index] = 'royalblue'  # Color diferente para el valor máximo

plt.pie(frecuencia[:10], labels=[str(i) for i in frecuencia.index[:10]],
        autopct='%1.1f%%', startangle=90, colors=colores)  # Colores personalizados
plt.title('Distribución de Proporciones de Valores (Top 10)', fontsize=16)
plt.axis('equal')  # Para que el pastel sea circular
plt.show()

# Gráfico de frecuencia acumulada (ojiva)
frecuencia_acumulada = frecuencia.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_acumulada.index, frecuencia_acumulada.values, marker='s', linestyle='-', color='forestgreen', markersize=8)  # Color más profesional
plt.title('Ojiva - Frecuencia Acumulada', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Frecuencia Acumulada', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.show()
```
