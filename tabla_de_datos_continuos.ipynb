```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de Seaborn
sns.set(style="whitegrid")

# Generar 500 datos aleatorios continuos entre 0 y 100
data = np.random.uniform(0, 100, 500)

# Crear un DataFrame para facilitar el manejo de datos
df = pd.DataFrame(data, columns=['Valores'])

# Mostrar la tabla de datos
print("Tabla de datos generados:")
print(df.describe().round(2))

# Grfico de frecuencia absoluta (histograma)
plt.figure(figsize=(10, 6))
plt.hist(df['Valores'], bins=5, color='lightcoral', alpha=0.7)
plt.title('Histograma de Conteo de Ocurrencias', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Conteo de Ocurrencias', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(bottom=0)
plt.show()

# Grfico de barras horizontales
frecuencia = df['Valores'].value_counts(bins=5).sort_index()
plt.figure(figsize=(10, 6))
frecuencia.plot(kind='barh', color='orange', alpha=0.75)
plt.title('Diagrama de Barras Horizontales', fontsize=16)
plt.xlabel('Nmero de Ocurrencias', fontsize=14)
plt.ylabel('Rangos de Valores', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Grfico de frecuencia relativa (polgono de frecuencia)
frecuencia_relativa = frecuencia / len(df)
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_relativa.index.mid, frecuencia_relativa.values, marker='o', linestyle='-', color='purple', markersize=8)
plt.title('Polgono de Proporcin de Datos', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Proporcin de Datos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.show()

# Diagrama de pastel
plt.figure(figsize=(10, 6))
plt.pie(frecuencia, labels=[f"{int(interval.left)}-{int(interval.right)}" for interval in frecuencia.index],
        autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Distribucin de Proporciones de Valores', fontsize=16)
plt.axis('equal')  # Para que el pastel sea circular
plt.show()

# Grfico de frecuencia acumulada (ojiva)
frecuencia_acumulada = frecuencia.cumsum()
frecuencia_acumulada = np.insert(frecuencia_acumulada.values, 0, 0)
valores_acumulados = np.insert(frecuencia.index.mid, 0, 0)

plt.figure(figsize=(10, 6))
plt.plot(valores_acumulados, frecuencia_acumulada, marker='o', linestyle='-', color='green', markersize=8)
plt.title('Ojiva - Frecuencia Acumulada', fontsize=16)
plt.xlabel('Rango de Valores', fontsize=14)
plt.ylabel('Frecuencia Acumulada', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.show()
```
